```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.optimize import curve_fit
```

```python
plt.rcParams['text.usetex'] = True
```

<h1>Two point function:</h1>

```python
tpf_df = pd.read_csv('two_point_func_data_40_zoom_ex.csv')
tpf = np.transpose(tpf_df.to_numpy())
print(tpf)
```

```python
def tpf_fit(x,a,b):
    return np.exp(-x/40)*x**b
```

```python
popt,pcov = curve_fit(tpf_fit,tpf[0][5:10],tpf[1][5:10])
```

```python
print(popt)
```

```python
plt.plot(tpf[0][:],tpf[1][:],label='data')
plt.plot(tpf[0][:],tpf_fit(tpf[0][:],*popt),label='fit')
plt.xlabel(r'$r$')
plt.ylabel(r'$G(r)$')
plt.title(r'$G(r)$ vs $r$ for $N=40$')
plt.grid(True,which='both')
plt.legend()
plt.show()
```

```python
res = linregress(np.log(tpf[0][0:8]),np.log(tpf[1][0:8]))
print(res.slope)
print(res.intercept)
```

```python
plt.plot(tpf[0][1:],np.gradient(np.log(tpf[1][1:]),np.log(tpf[0][1:])))
print(np.mean(np.gradient(np.log(tpf[1][1:]),np.log(tpf[0][1:]))))
```

<h1>Finite size scaling:</h1>

```python
def get_values(filename):
    val_df = pd.read_csv(filename)
    val = np.transpose(val_df.to_numpy())
    return val
```

```python
def scale(eta,nu,L,Tc_est,Ts,qs):
    return L**(1/nu)*(Ts-Tc_est)/Tc_est,L**(-eta/nu)*qs
```

```python
nu = 1
```

<h2>Calculating $T_c$ :</h2>

```python
def Tc_fit(x,a,b):
    return a - b / x
```

```python
sus_16_zoom = get_values('susceptibility_data_16_zoom.csv')
sus_25_zoom = get_values('susceptibility_data_25_zoom.csv')
sus_40_zoom = get_values('susceptibility_data_40_zoom.csv')
```

```python
plt.loglog(sus_16_zoom[0],sus_16_zoom[1],label='16')
plt.loglog(sus_25_zoom[0],sus_25_zoom[1],label='25')
plt.loglog(sus_40_zoom[0],sus_40_zoom[1],label='40')
plt.xlabel('$T$')
plt.ylabel(r'$\chi$')
plt.legend()
plt.grid(True,which='both')
plt.show()
```

```python
Ls = np.array([16,25,40])
Tcs = np.array([sus_16[0][np.argmax(sus_16[1])],sus_25[0][np.argmax(sus_25[1])],sus_40[0][np.argmax(sus_40[1])]])
popt,pcov = curve_fit(Tc_fit,Ls,Tcs)
```

```python
print(popt)
print(pcov)
```

```python
plt.plot(1/Ls,Tcs)
plt.xlabel(r'$L^{-1}$')
plt.ylabel('$T_c(L)$')
plt.grid()
plt.show()
```

```python
Tc_est = popt[0]
print('Estimated Tc is',Tc_est)
```

```python
Ls=[16,25,40]
Ti=list(range(10,17))
print(sus_16[0][Ti])
print(np.argmax(sus_16[1]))
print(np.argmax(sus_25[1]))
print(np.argmax(sus_40[1]))
lin_sus = linregress(np.log(Ls),np.log([sus_16[1][7],sus_25[1][10],sus_40[1][14]]))
print(lin_sus.slope)
for T in Ti:
    plt.loglog(Ls,[sus_16[1][T],sus_25[1][T],sus_40[1][T]],label=str(T))
plt.grid(True,which='both')
plt.legend()
plt.show()
```
